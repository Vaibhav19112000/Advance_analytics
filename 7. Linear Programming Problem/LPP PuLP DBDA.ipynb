```python
import pulp as p
```

# Problem 1

### Maximization

```python
Lp_prob = p.LpProblem('Problem1', p.LpMaximize)
```

### Decision Variables 

```python
x1 = p.LpVariable("x1", lowBound = 0) # Create a variable x >= 0
x2 = p.LpVariable("x2", lowBound = 0) # Create a variable y >= 0
```

### Objective Function 

```python
Lp_prob += 5*x1 + 7*x2
```

### Constraints 

```python
Lp_prob += 1*x1 + 0*x2 <= 6
Lp_prob += 2*x1 + 3*x2 <= 19
Lp_prob += 1*x1 + 1*x2 <= 8
```

```python
print(Lp_prob)
```

### Solving the LPP 

```python
status = Lp_prob.solve()
print(p.LpStatus[status])
```

### Solution 

```python
print("x1 =", p.value(x1)) 
print("x2 =", p.value(x2))
print("Objective = ", p.value(Lp_prob.objective))
```
